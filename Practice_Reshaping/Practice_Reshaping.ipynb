```python
import pandas as pd
```

# 1. โหลดไฟล์ที่ชื่อ olympics_cleaned.csv ขึ้นมาใน DataFrame ที่ชื่อ medals

```python
medals = pd.read_csv("olympics_cleaned.csv")
medals
```

# 2. โหลดไฟล์ชื่อ drinks.csv ขึ้นมาใน DataFrame ชื่อ drinks

```python
drinks = pd.read_csv("drinks.csv")
drinks
```

# 3. ใน drinks มีคอลัมน์ไหนที่ไม่มีข้อมูลบ้าง และไม่มีข้อมูลกี่ตำแหน่งในแต่ละคอลัมน์

```python
drinks.isnull().sum()
```

# 4. ลองเช็คว่าชื่อประเทศของตาราง medals ใดที่ไม่มีอยู่ในตาราง drinks

1. ลองใช้คำสั่ง isin() ในการเช็คว่ามีประเทศใดใน medals ไม่อยู่ใน drinks บ้าง

```python
medals.loc[medals['Country'].isin(drinks.country) == False]
```

# 5. แสดงชื่อประเทศของตาราง drinks ที่ไม่มีอยู่ในตาราง medals นิสิตจะเห็นความแตกต่างของชื่อประเทศในสองตารางนี้ว่าเรียกชื่อประเทศไม่เหมือนกัน ตอนทำงานจริงนั้นต้องปรับชื่อให้ตรงกัน (เช่น USA กับ United States) แต่ในการบ้านนี้ไม่ต้องทำ

```python
medals.loc[drinks['country'].isin(medals.Country) == False]
```

# 6. นำคอลัมน์ที่ชื่อ continent ของ drinks ไปต่อกับตาราง medals ตามชื่อประเทศของ medals

1. การที่จะ join ตารางได้นั้นต้องมีคอลัมน์ที่ตรงกันในทั้งสองตาราง ในที่นี้ให้ใช้ชื่อประเทศ
2. อย่าลืมว่าการ join ข้อมูลนั้นเราต้องการคงข้อมูลของประเทศที่มีในตาราง medals ไว้ ตารางที่ join แล้วจะมีจำนวนแถวเท่ากับตาราง medals

```python
medals_continent = medals.join(drinks.set_index('country').continent, on='Country', how='left', rsuffix='_drinks')
medals_continent
```

# 7 . ในตาราง medals จงแสดงชื่อประเทศที่ยังไม่มีข้อมูลในคอลัมน์ continent

```python
continent_null = medals_continent.loc[medals_continent['continent'].isnull()]
continent_null
```

# 8. นิสิตจะเห็นแนวทางการแก้ไขข้อมูลหรือรวมประเทศบางประเทศเข้าด้วยกัน แต่ในการบ้านนี้ยังไม่ต้องทำทั้งหมด แค่เติมข้อมูลทวีปที่ยังขาดหายไปในบางประเทศดังนี้ Australasia: OC, Canada/United States/Mexico: NA, Cuba/Jamaica: SA, Czech/Yugoslavia/Germany(ทั้งหมด)/Great Britain/Russia(ทั้งหมด)/Soviet: EU ประเทศที่เหลือกำหนดให้เป็น 'OT' ไป

1. ตารางนี้จะเป็นตารางแบบ wide ที่สมบูณ์ ควรเก็บไว้เพื่อใช้ในการคำนวณ

```python
for i in (continent_null.index):
    if medals_continent.iloc[i,0] == 'Australasia' :
        medals_continent.iloc[i,-1] = 'OC'
    elif medals_continent.iloc[i,0] in ['Canada','United States','Mexico']:
        medals_continent.iloc[i,-1] = 'NA'
    elif medals_continent.iloc[i,0] in ['Cuba','Jamaica']:
        medals_continent.iloc[i,-1] = 'SA'
    elif (medals_continent.iloc[i,0] in ['Czech','Yugoslavia','Great Britain'] or 
          'Soviet' in medals_continent.iloc[i,0] or 
          "Germany" in medals_continent.iloc[i,0] or 
          "Russia" in medals_continent.iloc[i,0]):
        medals_continent.iloc[i,-1] = 'EU'
    
    else:
        medals_continent.iloc[i,-1] = 'OT'

medals_continent.isnull().sum()
```

# 9. ตั้งให้ ชื่อทวีปเป็น index ชั้นนอกและชื่อประเทศเป็น index ชั้นใน

```python
medals_continent = medals_continent.set_index(['continent', 'Country'])
medals_continent
```

# 10. เรียงลำดับชื่อทวีปจากมากไปน้อยและในทวีปเรียงลำดับชื่อประเทศจากน้อยไปมาก

```python
medals_continent = medals_continent.sort_index(level = ['continent', 'Country'], ascending = [False, True])
medals_continent
```

# 11. จัดคอลัมน์ให้เป็นสองระดับ โดยระดับแรกคือฤดู และระดับที่สองคือจำนวนเกมส์และเหรียญที่ได้ (ดูตัวอย่างจากเอกสารการสอน) ดังนั้นตอนนี้ข้อมูลจะมีสองระดับทั้ง index (ทวีปและประเทศ) และ column (ฤดูและจำนวนเกมและเหรียญ)

```python
old_cols = medals_continent.columns
new_cols = [('Code',''),('Summer','Game'), ('Summer','Gold'), ('Summer','Silver'), ('Summer','Bronze'), 
            ('Winter','Game'), ('Winter','Gold'), ('Winter','Silver'), ('Winter','Bronze')]
new_medals_continent = medals_continent.copy()
new_medals_continent.columns = pd.MultiIndex.from_tuples(new_cols)
new_medals_continent
```

# 12. ย้ายระดับของฤดูมาเป็น index ชั้นในสุด

```python
new_medals_continent.swaplevel(0, 1, axis = 1)
```

# 13. ปรับตารางให้เป็นแบบ long โดยหนึ่งแถวจะมีคอลัมน์คือ ทวีป ประเทศ ฤดูแข่งข้น จำนวนเกม ประเภทเรียญ จำนวนเรียญ

```python
long_temp_df = new_medals_continent.melt(id_vars= [('Code', '')],
                                     value_vars= [('Summer',   'Game'),('Summer','Gold'),
                                                  ('Summer', 'Silver'),('Summer', 'Bronze'),('Winter',   'Game'),
                                                  ('Winter',   'Gold'),('Winter', 'Silver'),('Winter', 'Bronze')],
                                                  value_name = 'Count',
                                                  var_name = ['Season','Type'],
                                                  ignore_index= False)
long_temp_df
```

```python
long_temp_df = long_temp_df.reset_index()
long_temp_df.columns = ['continent','Country','Code','Season','Type','Count']
long_temp_df
```

```python
long_temp_df = long_temp_df.pivot(index= ['continent','Country','Code','Season'], columns='Type', values = 'Count')
long_temp_df = long_temp_df.reset_index()

long_df = long_temp_df.melt(id_vars= ['continent','Country','Season', 'Game'],
                                 value_vars= ['Bronze','Gold','Silver'],
                                 value_name = 'Count',
                                 var_name = ['Medal'])
                                
long_df = long_df.sort_values(by=['continent', 'Country'], ascending=[False, True], ignore_index= True)
long_df

```

```python
long_df.loc[long_df.Country == 'Austria']
```

# 14. แสดงจำนวนเหรียญทองโอลิมปิกฤดูร้อนรวมของแต่ละทวีปจากมากไปน้อย

```python
long_df.loc[(long_df.Medal == 'Gold') & (long_df.Season == 'Summer')].groupby('continent')['Count'].sum().sort_values(ascending=False)
```

# 15. นับว่าแต่ละทวีปมีกี่ประเทศ (ลองใช้คำสั่ง size/ ต่อจากการ groupby() )

```python
new_medals_continent.groupby('continent').size()
```

# 16. แสดงตารางรวมเหรียญฤดูร้อนแต่ละประเภทของทุกประเทศในทวีปเดียวกัน แสดงผลแยกตามทวีป

```python
medals_continent.reset_index().groupby('continent')[['SummerGold', 'SummerSilver', 'SummerBronze']].sum()
```

# 17. แสดงตารางสรุปรวมเหรียญแต่ละประเภทของทุกประเทศแยกตามฤดูแข่งขัน

```python
pv = long_df.pivot_table(index = 'Medal',
               columns = 'Season',
               values = 'Count',
               aggfunc = 'sum')
pv
```

# 18. หาค่าเฉลี่ยของจำนวนเหรียญทองฤดูร้อนต่อประเทศที่เข้าร่วมแข่งขัน แยกตามทวีป

```python
long_df.loc[(long_df.Medal == 'Gold') & (long_df.Season == 'Summer')].groupby('continent')['Count'].mean()
```

# 19. หาจำนวนประเทศที่เคยเข้าร่วมการแข่งขัน โอลิมปิค แยกตามทวีปและฤดูแข่งขัน แสดงในตารางเดียวกัน

```python
long_df[(long_df.Game > 0) & (long_df.Medal == 'Gold')].pivot_table(index='continent',columns= 'Season', values='Country', aggfunc='count', fill_value=0)
```

# 20. จำนวนเหรียญรวม (ฤดูร้อนกับฤดูหนาว) แยกตามทวีปและฤดูแข่งขัน
