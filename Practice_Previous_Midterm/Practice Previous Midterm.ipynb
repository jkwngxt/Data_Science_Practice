# 15.คัดเลือก 7 สายการบิน (Operator) ดังนี้ออกมา All Nippon Airways, EVA Air, Scandinavian Airlines (SAS), Singapore Airlines, Swissair, Thai Airways, United Air Lines (ชื่อทั้งหมดอยู่ในไฟล์star_alliance_7.csv) ตั้งชื่อว่า star_7 แล้วแสดงข้อมูลในตารางนี้

```python
star_alliance_7 = pd.read_csv('star_alliance_7.csv')
name = star_alliance_7['Star Alliance']
star_alliance_7
```

```python
star_7 = df.loc[df.Operator.isin(name)]
star_7.head(5)
```

# 16.สรุปรวมจำนวนอุบัติเหตุที่เกิดขึ้นของแต่ละสายการบินใน star_7 *ระวัง: เช็คว่ามี 7 สายการบิน

```python
star_7['Operator'].value_counts().reindex(name, fill_value=0)
```

# 17.นำอุบัติเหตุของการบินไทย (Thai Airways) มาเรียงลำดับจากล่าสุดย้อนกลับไป

```python
star_7['Timeline'] = (star_7['Date']+ " " + (star_7['Time']))
star_7['Timeline'] = pd.to_datetime(star_7['Timeline'])
```

```python
star_7.loc[star_7.Operator == 'Thai Airways'].sort_values(['Timeline'], ascending = [False])
```

# 18.แสดงข้อมูลของ 7 สายการบินเรียงตาม Operator จากน้อยไปมากและในแต่ละ Operator เรียงจากอุบัติเหตุที่เกิดขึ้นก่อนมายังครั้งล่าสุด

```python
star_7.sort_values(['Operator','Timeline'], ascending = [True, True])
```

# 19.คัดเลือกอุบัติเหตุของสายการบิน Thai Airways ในทศวรรษที่ 90 (ปี 1990–1999) ออกมา

```python
star_7.loc[(star_7.Operator == 'Thai Airways') &
           star_7.Timeline.dt.year.isin(range(1990,2000))]
```

# 20.ในตาราง star_7 สกัดคำแรกของคอลัมน์ 'AC Type' ออกมาเป็นชื่อบริษัทที่สร้างเครื่องบินจากนั้นนับจำนวนเครื่องบินที่ตกแยกตามสายการบินและบริษัทเครื่องบิน (ไม่ต้องสนใจข้อมูลที่อาจผิดพลาด)

```python
star_7['Company'] = star_7['AC Type'].str.split(" ").str.get(0)
star_7
```

# 21.นับจำนวนคนตายของ 7 สายการบินแยกตามทศวรรษ (1930, 1940, 1950, ... , 2000) และสายการบิน

```python
temp = star_7.copy()
temp['Century'] = temp.Timeline.dt.year // 10 * 10
temp.loc[temp.Century.isin(range(1930, 2001))].pivot_table(index = 'Operator', columns = 'Century', values = 'Fatalities', aggfunc = 'sum', fill_value = 0)
```

# 22.เปลี่ยนคอลัมน์ Operator ของ star_7 ให้กลายเป็น category แบบเรียงลำดับโดยลำดับคือลำดับตามตัวอักษร (ตามข้อ 15)

```python
star_7['Operator'] = pd.Categorical(star_7['Operator'], ordered = True,
                                    categories = ['All Nippon Airways','EVA Air','Scandinavian Airlines (SAS)',
                                                  'Singapore Airlines','Swissair','Thai Airways','United Air Lines'])
star_7['Operator']
```

# 23.แสดงแผนภาพ Histogram และ Rug plot ของ Fatalities Percent ของอุบัติเหตุจาก 7 สายการบิน

```python
sns.displot(star_7['Fatalities Percent'], height=3, rug=True)
```

# 24.แสดงกราฟแท่งแสดงจำนวนคนตายรวมของ 7 สายการบิน

```python
sns.catplot(data=star_7, x='Operator', 
            y='Fatalities', estimator='sum', kind='bar',
            errorbar=None)
```

# 25.จากตาราง star_7 แสดงแผนภาพแบบ scatterplot แสดงจำนวนครั้งที่เครื่องบินตกเทียบกับจำนวนคนตายเฉลี่ยต่ออุบัติเหตุหนึ่งครั้งและแยกสีตามสายการบินขนาดของจุดแปรผันตามจำนวนคนตายรวมของสายการบิน

```python
temp = star_7.groupby('Operator')['Fatalities'].agg(['sum','mean','count']).reset_index()
```

```python
sns.relplot(data = temp, x = 'count', y = 'mean', hue = 'Operator', size = 'sum')
```
