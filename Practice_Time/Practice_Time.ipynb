```python
import pandas as pd
```

# 1.ปิดไฟล์ italy_earthquakes_from_2016-08-24_to_2016-11-30.csv

```python
df = pd.read_csv('italy_earthquakes_from_2016-08-24_to_2016-11-30.csv')
df.head(5)
```

# 2. ปรับคอลัมน์ Time ให้เป็น Timestamp และตั้งให้เป็น index

```python
df.Time = pd.to_datetime(df.Time)
df = df.set_index('Time').sort_index(ascending = True)
df.head(5)
```

# 3. เวลาที่เกิดแผ่นดินไหวแรงที่สุดคือวันเวลาไหน

```python
day_max = df.loc[df.Magnitude == df.Magnitude.max()].index[0]
day_max
```

# 4. ในวันที่เกิดแผ่นดินไหวแรงที่สุดมีแผ่นดินไหวกี่ครั้ง

```python
day_max.strftime('%Y-%m-%d')
```

```python
df.loc['2016-10-30'].shape[0]
```

# 5. แสดงข้อมูลตั้งแต่ 7 ถึง 9 โมงเช้าในวันที่เกิดแผ่นดินไหวแรงที่สุด

# 6. แผ่นดินไหวที่แรงที่สุดกับแรงเป็นที่ 2 ห่างกันเท่าไหร่ตอบเป็นวัน-ชั่วโมง-นาที

# 7. คำนวณความแรงของแผ่นดินไหวที่เปลี่ยนไปในแต่ละครั้งของข้อมูลในข้อ 5.
หมายเหตุ: ใช้คำสั่ง shift() ช่วยเลื่อนแถวของคอลัมน์ Magnitude เพื่อคำนวณความแตกต่าง

```python
temp = df.copy()
temp['DiffLastMagnitude'] = temp.Magnitude.diff(1)
temp
```

# 8. มีแผ่นดินไหวเกิดขึ้นกี่ครั้งในแต่ละวัน

```python
temp.resample('D').size()
```

# 9. ถ้าแยกเวลาเป็นช่วงช่วงละหนึ่งชั่วโมงมีกี่ช่วงที่ไม่มีแผ่นดินไหวเกิดขึ้นเลยในชั่วโมงนั้น

```python
a = temp.resample('H').size().to_frame()
a.columns = ['Count']
a.loc[a.Count == 0].shape[0]
```

# 10. ถ้าแยกเวลาเป็นช่วงช่วงละ 5 นาทีช่วงเวลาไหนมีค่าเฉลี่ยความรุนแรงของแผ่นดินไหวสูงที่สุดและค่าเฉลี่ยนั้นมีค่าเท่าไหร่

```python
b = temp.resample('5 min')['Magnitude'].mean().to_period().to_frame().fillna(0)
b.columns = ['Mean']
b.loc[b.Mean == b.Mean.max()]
```

# 11. ระยะเวลาที่ไม่มีแผ่นดินไหวเลยนานที่สุดคือเท่าไหร่

```python
e11 = df.reset_index().Time.to_frame()
e11.columns = ['Stop']
e11['Start'] = e11.Stop.shift(1)
e11['Diff'] = e11.Stop - e11.Start 
e11.Diff.max()
```

# 12. ระยะเวลาที่ไม่มีแผ่นดินไหวเลยนานที่สุดคือช่วงไหนระบุเป็นเวลาเริ่มต้นและเวลาจบ

```python
e11.loc[e11.Diff == e11.Diff.max()]
```

# 13. สมมติว่าเรานิยามแผ่นดินไหวที่รุนแรงนั้นต้องมี Magnitude มากกว่าการไหวครั้งก่อนและครั้งหลังอย่างน้อย 2.0 จงคัดเลือกเหตุการณ์แผ่นดินไหวที่รุนแรงออกมา

```python
temp['DiffAfterMagnitude'] = temp.Magnitude.diff(-1)
c = temp.loc[(temp['DiffAfterMagnitude'] >= 2) & (temp['DiffLastMagnitude'] >= 2)]
c
```

# 14. จากตารางข้อ 13. ให้เพิ่มคอลัมน์ใหม่ชื่อ hour_count แสดงจำนวนแผ่นดินไหวในชั่วโมงที่มีเหตุการณ์แผ่นดินไหวรุนแรงด้วย

```python
c['hour_count'] = 0
for i in range(c.shape[0]) :
    date = c.reset_index().iloc[i,0].strftime('%Y-%m-%d %H')
    c.iloc[i,-1] = df.loc[date].shape[0]
c
```

# 15. แสดงจำนวนแผ่นดินไหวแยกตาม 24 ชั่วโมง (เอาจำนวนแผ่นดินไหวของแต่ละชั่วโมง 0-23 ของทุกวันรวมกันแสดงผลตามชั่วโมง)

```python
dd = df.resample('H').size().to_frame().reset_index()
dd.columns = ['Date', 'Count']
dd['Hour'] = dd['Date'].dt.strftime("%H")
dd['Date'] = dd['Date'].dt.date
dd.pivot_table(index = 'Date', columns = 'Hour', values = 'Count', aggfunc = 'sum', fill_value= 0)
```

# 16. คำนวณค่าเฉลี่ยของความรุนแรงแผ่นดินไหวที่มากที่สุดในช่วง 3 วันโดยหาความแรงของแผ่นดินไหวที่มากที่สุดในแต่ละวันก่อนแล้วใช้ sliding window หาค่าเฉลี่ยของทุกๆ 3 วัน

```python
rd = df.Magnitude.resample('D').max().rolling(3)
```

```python
for i in rd :
    print(i, i.mean())
    print('mean =',i.mean())
    print('---------------------------------')
```
